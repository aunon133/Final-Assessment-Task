```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
from datetime import datetime, timedelta
```

```python
chromedriver_path = 'C:\\Users\\aunon\\Downloads\\chromedriver-win64\\chromedriver.exe'

service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service)

url = "https://bangla.dhakatribune.com/economy"
driver.get(url)
time.sleep(3) 
```

```python
article_urls = set()

articles = driver.find_elements(By.XPATH, '//a[contains(@href, "/economy/")]')
for article in articles:
    href = article.get_attribute('href')
    article_urls.add(href)

driver.quit()
print(f"Extracted URLs: {article_urls}")
```

```python
def scrape_article_data(url):
    chromedriver_path = 'C:\\Users\\aunon\\Downloads\\chromedriver-win64\\chromedriver.exe'
    
    service = Service(chromedriver_path)
    driver = webdriver.Chrome(service=service)
    
    driver.get(url)
    time.sleep(2)

    title = driver.find_element(By.TAG_NAME, 'h1').text
    
    sup_text = driver.find_element(By.CSS_SELECTOR, 'div.content_highlights.jw_detail_content_holder.content.mb16 strong').text
    meta_description = driver.find_element(By.NAME, 'description').get_attribute('content')
    
    alignfull_paragraphs = driver.find_elements(By.CLASS_NAME, 'alignfull')
    contents = ' '.join([paragraph.text for paragraph in alignfull_paragraphs])
    
    published_time_raw = driver.find_element(By.CLASS_NAME, 'tts_time.published_time').text
    published_time = published_time_raw.replace("প্রকাশ :", "").strip()
    
    modified_time_raw = driver.find_element(By.CLASS_NAME, 'modified_time').text
    modified_time = modified_time_raw.replace("আপডেট :", "").strip() 
    
    image_url = driver.find_element(By.CSS_SELECTOR, 'div.featured_image_outer img').get_attribute('src')

    keywords = []
    keyword_elements = driver.find_elements(By.CSS_SELECTOR, 'div.topic_list a')
    for element in keyword_elements:
        keyword = element.text.strip()
        keywords.append(keyword)


    article_data = {
        "url": url,
        "title": title,
        "support_text": sup_text,
        "news_type": "Economics",
        "meta_description": meta_description,
        "content": contents,
        "published_time": published_time,
        "modified_time": modified_time,
        "Image": image_url,
        "keywords": keywords,
        "views": 0,
        "rating": 0,
        
    }
    
    driver.quit()
    return article_data
```

```python
for article_url in article_urls:
    article_data = scrape_article_data(article_url)
    print(article_data)
```

```python
import json

def analyze_article(url):
    article_data = scrape_article_data(url)

    content = article_data['content']
    sentiment = get_sentiment(content)
    importance_score = get_importance_score(content)
    international_perspective = get_international_perspective(content)

    article_data["sentiment_analysis"] = sentiment
    article_data["importance_score"] = importance_score
    article_data["international"] = international_perspective

    return article_data

all_articles_data = []

for article_url in article_urls:
    article_data = analyze_article(article_url)
    all_articles_data.append(article_data)

with open('article_analysis.json', 'w', encoding='utf-8') as f:
    json.dump(all_articles_data, f, ensure_ascii=False, indent=4)
```
